# Tree

### Binary Tree [source](http://www.geeksforgeeks.org/data-structures/#AdvancedDataStructure)

A tree whose elements have at most 2 children is called a binary tree. Since each element in a binary tree can have only 2 children, we typically name them the left and right child.

![img](https://www.geeksforgeeks.org/wp-content/uploads/binary-tree-to-DLL.png)

```python
class Node(object):
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

```python
class BinaryTree(object):
    def __init__(self, root):
        self.root = Node(root)

        
    '''search helper recursive function'''   
    def preorder_search(self, start, find_val):
        if start:
            print(start.value)
            if start.value == find_val:
                return True
            else:
                return self.preorder_search(start.left, find_val) \
                     or self.preorder_search(start.right, find_val)
        return False
    
    
    '''return true if value exist in tree'''
    def search(self, find_val):
        return self.preorder_search(self.root, find_val)
```

```python
# Set up tree
tree = BinaryTree(1)
tree.root.left = Node(2)
tree.root.right = Node(3)
tree.root.left.left = Node(4)
tree.root.left.right = Node(5)
```

```python
# Test search
# Should be True
print (tree.search(4))
```

```python
# Should be False
print (tree.search(6))
```
